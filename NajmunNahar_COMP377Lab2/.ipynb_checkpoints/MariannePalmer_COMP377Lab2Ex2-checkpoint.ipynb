```python
import tensorflow as tf
import pandas as pd
from sklearn import svm
```

```python
# https://www.educative.io/edpresso/what-is-the-mnist-dataset-in-keras
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
print('Shape of X_train: ', X_train.shape)
print('Shape of y_train: ', y_train.shape)
print('Shape of X_test: ', X_test.shape)
print('Shape of y_test: ', y_test.shape)
```

```python
n_samples, nx, ny = X_train.shape   # Grab tuple values from shape attribute
```

```python
type(X_train)
```

```python
type(y_train)
```

```python
X_train= X_train.reshape(n_samples, nx * ny)
```

```python
X_test = X_test.reshape(len(X_test), nx * ny)
```

```python
# Showing the data shapes again
print('Shape of X_train: ', X_train.shape)
print('Shape of y_train: ', y_train.shape)
print('Shape of X_test: ', X_test.shape)
print('Shape of y_test: ', y_test.shape)
```

```python
# Convert to dataframe
X_train_df = pd.DataFrame(X_train)
X_test_df = pd.DataFrame(X_test)
```

```python
X_train_df
```

```python
# Modifies the existing dataframe to top 60 rows for training and top 10 rows for test
X_train_df.drop(X_train_df.index[60 : ], axis='index', inplace=True)
X_test_df.drop(X_test_df.index[10 : ], axis='index', inplace=True)
```

```python
X_train_df
```

```python
# Use top 20 rows and converting data to Series
y_train_ser = pd.Series( y_train[:60] )
y_test_ser = pd.Series( y_test[:10] )
```

```python
# Create a SVM's Support Vector Classifier model 
model1 = svm.SVC(kernel='linear', C = 1)
```

```python
model1.fit(X_train_df, y_train_ser ) # Train/fit the model
```

```python
y_pred = model1.predict(X_test_df) # Testing step
```

```python
pd.Series(y_pred)
```

```python
y_test_ser # compare with predicted values
```

```python
model1.score(X_test_df, y_test_ser) # the mean accuracy on the given test data and labels
```

```python
# imports for assessing accuracy
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
# Confusion matrix to compare predicted and true values of y
cm = confusion_matrix(y_test_ser, y_pred, labels=model1.classes_) 
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model1.classes_)
disp.plot()
plt.show()
```

```python
# Number of correctly classified samples
print(accuracy_score(y_test_ser, y_pred, normalize=False))
```

```python
# Fraction of correctly classified samples
print(accuracy_score(y_test_ser, y_pred))
```
