```python
#import librarires
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn import linear_model
from sklearn.neural_network import MLPClassifier
```

```python
#Reading dataset
df = pd.read_csv('student.cleaned.data.csv')
df
```

```python
#Taking the features into account (traveltime, studytime, failures, famrel, freetime, goout, health)
X = df[['traveltime', 'studytime', 'failures', 'famrel', 'freetime', 'goout', 'health']]
```

```python
y = np.where(df['G3'] < 10, 0, 1)
```

```python
#Initialize Logistic regression model
logistic=linear_model.LogisticRegression()
#Fitting the data
logistic.fit(X,y)
```

```python
#Initialize MLPClassifier
mlp=MLPClassifier(max_iter=300, random_state=1)
#Fitting the data
mlp.fit(X,y)
```

```python
#Evaluate Accuracy of Logisitic Model
score=logistic.score(X,y)
print('Accuracy of Logistic Model: ',score)
```

```python
#Evaluate Accuracy of Model using MLP
score2=mlp.score(X,y)
print('Accuracy of Model using MLP classifier: ',score2)
```

```python
#Here the accuracy score computed for MLP classifier is a little higher than the logistic model.This can conclude 
#MLP classifier here provides better performance than Logisitic
```
